**Categorical** **Plots**

This plot deals with categorical variables and how they can be visualized.

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
tips.describe()
```

```python
# change the estimator from mean to standard devaition
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# A countplot basically counts the categories and returns a count of their occurrences. 
sns.countplot(x='sex', data=tips)
```

A boxplot is sometimes known as the box and whisker plot.It shows the distribution of the quantitative data that represents the comparisons between variables. boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution i.e. the dots indicating the presence of outliers.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

Violinplot

It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the
data distribution. 

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

Stripplot

It basically creates a scatter plot based on the category.

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True,hue='sex',split=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```
